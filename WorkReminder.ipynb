```python

import cv2
import easyocr
from datetime import date, datetime
import smtplib
```

```python
import pandas as pd
```

```python
image = cv2.imread("/content/Shiftplan.jpg")
```

```python
reader = easyocr.Reader(['de'], gpu=True)
```

```python
results = reader.readtext("/content/Shiftplan.jpg", text_threshold=0.5, width_ths=0)
```

```python
print(results)
```

```python
text_df = pd.DataFrame(results, columns=['bbox', 'text', 'confidence'])
```

```python

text_df2 = text_df[text_df['confidence'] >0.5]['text']
```

```python
text_df2
```

```python
raw = text_df2.str.replace('-', '')
```

```python
working = raw.to_list()
```

```python
group = [working[i:i+4] for i in range(0, len(working), 4)]
```

```python
newdf = pd.DataFrame(group, columns=['Date', 'Day', 'Start', 'End'])
```

```python
newdf
```

```python
newdf['Date'] = pd.to_datetime(newdf['Date'], format="mixed", dayfirst=True)
```

```python
newdf
```
